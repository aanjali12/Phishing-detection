```python
X = d1.iloc [:,0:31]  # X is independent variable
y = d1.iloc [:,31]   #dependent variable
```

```python
from sklearn.preprocessing import StandardScaler
independent_scalar = StandardScaler()
X = independent_scalar.fit_transform (X) #fit and transform
d2 = independent_scalar.transform (d2) # only transform
#print(X)
#==============================================================================
# Encode the categorial data. So now instead of character values we will have
# corresponding numerical values
#==============================================================================
X
```

```python
#principal component analysis is a dimensionality reduction technique 
#i.e. given 10 input features,
#it will produce a smaller number of independent features., 
#it maximizes the variances. 

from sklearn.decomposition import PCA

pc = PCA(n_components=2)
X=pc.fit_transform(X)
d2= pc.transform(d2)
```

```python
from sklearn.ensemble import RandomForestClassifier
RFclassifier = RandomForestClassifier(n_estimators=10, random_state=42, criterion = 'entropy')


RFclassifier.fit(X,y) # training



prediction = RFclassifier.predict(d2)# testing 


print(prediction)
```

```python
import xgboost as xgb
classifier = xgb.XGBClassifier()
classifier.fit(X,y)
predicted = classifier.predict(d2)
predicted
```

```python
df=pd.read_csv("Phising_Testing_Dataset.csv")
df=df['key'].tolist()
```

```python
result = pd.DataFrame(list(zip(df, predicted)),
               columns =['Key', 'Result'])
result
```
